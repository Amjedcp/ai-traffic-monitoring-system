```python
import cv2
import csv
from datetime import datetime
from ultralytics import YOLO

# ================= CONFIG =================
VIDEO_PATH = "vehicle-counting.mp4"
OUTPUT_PATH = "output.mp4"
CSV_PATH = "traffic_data.csv"

MODEL_PATH = "yolov8n.pt"
OVERSPEED_LIMIT = 40
PIXEL_TO_METER = 0.04

# ================= INIT =================
model = YOLO(MODEL_PATH)
model.to("cuda")

cap = cv2.VideoCapture(VIDEO_PATH)

width = int(cap.get(3))
height = int(cap.get(4))
fps = cap.get(cv2.CAP_PROP_FPS) or 30

out = cv2.VideoWriter(
    OUTPUT_PATH,
    cv2.VideoWriter_fourcc(*'mp4v'),
    fps,
    (width, height)
)

line_y = int(height * 0.65)

# ================= DATA =================
up_count = 0
down_count = 0

vehicle_count = {"car": 0, "bus": 0, "truck": 0, "motorbike": 0}

track_history = {}
counted_ids = set()
speed_memory = {}
saved_overspeed = set()

MAX_HISTORY = 5

# ================= CSV =================
csv_file = open(CSV_PATH, mode="w", newline="")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["ID", "Type", "Speed(km/h)", "Direction", "Time"])

# ================= FUNCTIONS =================
def calculate_speed(history, fps):
    if len(history) < 2:
        return 0

    (x_old, y_old) = history[0]
    (x_new, y_new) = history[-1]

    dist_pixels = ((x_new - x_old)**2 + (y_new - y_old)**2) ** 0.5
    dist_meters = dist_pixels * PIXEL_TO_METER

    time_sec = len(history) / fps
    if time_sec <= 0:
        return 0

    return (dist_meters / time_sec) * 3.6


def smooth_speed(track_id, new_speed):
    if track_id in speed_memory:
        return 0.7 * speed_memory[track_id] + 0.3 * new_speed
    return new_speed


# ================= MAIN LOOP =================
while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(
        frame,
        persist=True,
        tracker="bytetrack.yaml",
        conf=0.35,
        iou=0.5,
        device=0,
        verbose=False
    )[0]

    if results.boxes is not None and results.boxes.id is not None:

        boxes = results.boxes.xyxy.cpu().numpy()
        ids = results.boxes.id.cpu().numpy()
        classes = results.boxes.cls.cpu().numpy()

        for box, track_id, cls in zip(boxes, ids, classes):

            label = model.names[int(cls)]
            if label not in vehicle_count:
                continue

            x1, y1, x2, y2 = map(int, box)
            track_id = int(track_id)

            cx = int((x1 + x2) / 2)
            cy = int((y1 + y2) / 2)

            # TRACK HISTORY
            if track_id not in track_history:
                track_history[track_id] = []

            track_history[track_id].append((cx, cy))

            if len(track_history[track_id]) > MAX_HISTORY:
                track_history[track_id].pop(0)

            # SPEED
            raw_speed = calculate_speed(track_history[track_id], fps)
            speed_kmh = smooth_speed(track_id, raw_speed)
            speed_memory[track_id] = speed_kmh

            # COLOR + ALERT
            color = (0, 200, 0)
            alert = ""

            if speed_kmh > OVERSPEED_LIMIT:
                color = (0, 0, 255)
                alert = "ALERT"
            elif speed_kmh > 20:
                color = (0, 255, 255)

            # COUNTING
            direction = ""
            current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            if len(track_history[track_id]) >= 2:
                prev_y = track_history[track_id][-2][1]
                curr_y = track_history[track_id][-1][1]

                if prev_y < line_y <= curr_y:
                    if track_id not in counted_ids:
                        down_count += 1
                        vehicle_count[label] += 1
                        counted_ids.add(track_id)

                        csv_writer.writerow(
                            [track_id, label, int(speed_kmh), "DOWN", current_time]
                        )
                    direction = "↓"

                elif prev_y > line_y >= curr_y:
                    if track_id not in counted_ids:
                        up_count += 1
                        vehicle_count[label] += 1
                        counted_ids.add(track_id)

                        csv_writer.writerow(
                            [track_id, label, int(speed_kmh), "UP", current_time]
                        )
                    direction = "↑"

            # ===== PROFESSIONAL VEHICLE TEXT =====
            text = f"{label.upper()} | {int(speed_kmh)} km/h"
            if alert:
                text += " | ALERT"

            # Shadow (bold)
            cv2.putText(frame, text, (x1, y1 - 12),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,0,0), 5)

            # Main text
            cv2.putText(frame, text, (x1, y1 - 12),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

    # ===== BLUE LINE =====
    cv2.line(frame, (0, line_y), (width, line_y), (255, 0, 0), 3)

    # ===== UP/DOWN TEXT (SHIFT RIGHT) =====
    shift = 90

    cv2.putText(frame, f"UP: {up_count}",
                (width // 2 - 80 + shift, line_y - 20),
                cv2.FONT_HERSHEY_SIMPLEX, 1,
                (255,255,255), 3)

    cv2.putText(frame, f"DOWN: {down_count}",
                (width // 2 - 100 + shift, line_y + 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1,
                (255,255,255), 3)

    # ===== ANALYTICS PANEL =====
    y_offset = 100
    for v_type, count in vehicle_count.items():
        cv2.putText(frame, f"{v_type.upper()}: {count}",
                    (20, y_offset),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.75,
                    (255,255,255), 2)
        y_offset += 30

    # ===== PROFESSIONAL TIME DISPLAY =====
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    cv2.rectangle(frame, (width - 340, 10), (width - 10, 55), (0,0,0), -1)

    cv2.putText(frame, current_time,
                (width - 330, 42),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.85,
                (0,0,0), 4)

    cv2.putText(frame, current_time,
                (width - 330, 42),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.85,
                (255,255,255), 2)

    out.write(frame)

# ================= CLEANUP =================
cap.release()
out.release()
csv_file.close()

print("✅ FINAL ULTRA PROFESSIONAL SYSTEM READY 🚀")
```
